1. For the suburb Altona, it is postulated that a typical property sells for $800,000. Use the data at hand to test this assumption. Is the typical property price really $800,000 or has it increased? Use a significance level of 5%.

```python
t_stat, p_value = ttest_1samp(df.loc[df['Suburb'] == 'Altona']['Price'], 800000)
print(t_stat, p_value)
```

Since P value is less tha sigificance value, we dont have enough evidence to conclude that the typical property price really has increased $800,000 

2. For the year 2016, is there any difference in prices of properties sold in the summer months vs winter months? Consider months from October till March as winter months and rest as summer months. Use a significance level of 5%.

```python
df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
df_2016 = df[df['Year'] == 2016]
df_2016
```

```python
summer_months = [4, 5, 6, 7, 8, 9]
winter_months = [10, 11, 12, 1, 2, 3]
```

```python
summer_data = df_2016[df_2016['Month'].isin(summer_months)]
winter_data = df_2016[df_2016['Month'].isin(winter_months)]
```

```python
t_stat, p_value = ttest_ind(summer_data['Price'], winter_data['Price'], equal_var=False)
print(t_stat, p_value)
```

he p value is significantlly less than significance level hence we can conclude that there is a statistically significant difference in property prices between the summer and winter months in 2016.

3. For the suburb of Abbotsford, what is the probability that out of 10 properties sold, 3 will not have car parking? Use the column car in the dataset. Round off your answer to 3 decimal places

```python
abbotsford_data = df[df['Suburb'] == 'Abbotsford']
no_car_proportion = (abbotsford_data['Car'] == 0).mean()
probability = binom.pmf(3, 10, no_car_proportion)
no_car_proportion, round(probability, 3)
```

The probability that out of 10 properties sold, 3 will not have car parking is 0.26
